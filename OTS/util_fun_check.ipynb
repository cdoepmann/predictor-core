```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
import pdb
from scipy.interpolate import interp1d
from optimal_traffic_scheduler import interpol_nd
```

```python
# Mimic input/output shape as it is used in latency_adaption
# x is the "time vector" with delays: [0.1, 0.2, 0.3] s 
delay = np.array([[0.1, 0.2, 0.3]])
x = np.atleast_3d(np.arange(-1, 10).reshape(-1,1) + delay)
# y is the corresponding output.
y = np.sin(x)
print('shape of x: {} (time_steps x elements x 1)'.format(x.shape))
print('shape of y: {} (time_steps x elements x 1)'.format(y.shape))
# Note: It is important that time_steps and number of elements for x and y are matching 
# or that x only has one element.
```

```python
x_new = np.arange(10).reshape(-1,1,1)+np.array([0.0, 0.0, 0.0]).reshape(1,-1,1)
y_new = interpol_nd(x,y,x_new)

plt.step(x.squeeze(),y.squeeze(),where='post')
plt.gca().set_prop_cycle(None)
plt.step(x_new.squeeze(),y_new.squeeze(),'*--', linestyle='--', where='post')
plt.legend()
plt.show()
```
